# Deezer API — Exploration

```python
import requests

def preview_playlist(playlist_id: str, n: int = 5):
    """Affiche les n premiers sons d'une playlist Deezer publique."""
    pl = requests.get(f"https://api.deezer.com/playlist/{playlist_id}").json()
    print(f"Playlist : {pl['title']}  —  {pl['nb_tracks']} tracks  —  par {pl['creator']['name']}")
    print("-" * 50)

    tracks = requests.get(f"https://api.deezer.com/playlist/{playlist_id}/tracks").json().get("data", [])
    for i, t in enumerate(tracks[:n], 1):
        duration = f"{t['duration'] // 60}:{t['duration'] % 60:02d}"
        print(f"{i}. {t['artist']['name']} — {t['title']}  [{duration}]")
```

```python
preview_playlist("15147697423")
```

```python
def list_playlists(user_id: str):
    """Liste les playlists publiques d'un utilisateur dont le nom commence par 'W -', avec pagination."""
    playlists = []
    url = f"https://api.deezer.com/user/{user_id}/playlists?limit=50&index=0"
    while url:
        resp = requests.get(url).json()
        playlists.extend(resp.get("data", []))
        url = resp.get("next")  # None si dernière page

    filtered = [pl for pl in playlists if pl["title"].startswith("W -")]
    print(f"{len(filtered)} playlists 'W -' trouvées (sur {len(playlists)} au total)")
    print("-" * 50)
    for pl in filtered:
        print(f"[{pl['id']}]  {pl['title']}  —  {pl['nb_tracks']} tracks")

list_playlists("656772321")
```
